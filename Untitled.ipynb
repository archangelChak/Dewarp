```python
from processing.postprocessing import find_grid
from processing.preprocessing import Smooth
from PIL import Image
```

```python
mask = np.asarray(mask[0].transpose(0,1).transpose(1,2))
image = np.asarray(image[0])
```

```python
grid = find_grid(Smooth(np.asarray(mask).squeeze()), np.asarray(mask).squeeze())
grid
```

```python
from dewarp.dewarp import show_result, get_pred_mask
```

```python
from functools import reduce
import cv2
```

```python
image[0].shape
```

```python
x=[]
y=[]
for j in range(11):
    x+= [[i[0] + q for q in range(-10,10,1)] for i in grid[j]]
    y+= [[i[1] + q for q in range(-10,10,1)] for i in grid[j]]
    x+= [[i[0] + q for q in range(-10,10,1)] for i in grid[j]]
    y+= [[i[1] + q for q in range(10,-10,-1)] for i in grid[j]]
x = reduce(lambda a,b: a+b,x)
y = reduce(lambda a,b: a+b,y)
grid_img = cv2.cvtColor(np.asarray(image[0]*255,dtype=np.uint8),cv2.COLOR_GRAY2RGB)
grid_img[x, y] = [255,0,0]
display(Image.fromarray(grid_img))
```
